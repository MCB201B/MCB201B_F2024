Create a line plot that contains the following:
<ul>
    <li>X-axis label</li>

    <li>Y-axis with 0 set to the bottom left corner</li>
</ul>

```python
#There are multiple ways to approach this question.
#Convert to datetime and reassign back to original DataFrame
COVID_datetime = pd.to_datetime(COVID_all['end_date'])
COVID_all = COVID_all.assign(end_date=COVID_datetime)

#Filter for NY state
NY_data = COVID_all.loc[COVID_all['state'] == 'NY']

#Then filter the date range
NY_data_filter_lower = NY_data.loc[NY_data['end_date'] > pd.to_datetime('02/29/2020')]
NY_data_filtered = NY_data_filter_lower.loc[NY_data_filter_lower['end_date'] < pd.to_datetime('01/01/2021')]

#Set up plot attributes
plt.ylabel('Number of new weekly cases')
plt.xlabel('Date')
plt.xticks(rotation=30, ha='right')
plt.ylim(0, 55000)

#Plot the data by providing the X values and Y values as arguments
plt.plot(NY_data_filtered['end_date'],
         NY_data_filtered['new_cases'])
```
